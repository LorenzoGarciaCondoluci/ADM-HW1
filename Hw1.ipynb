```python
#String Formatting

def print_formatted(n):
    width = len("{0:b}".format(n))
    for i in range(1, n+1):
        n_decimal = "{:d}".format(i)
        n_octal = "{:o}".format(i)
        n_hex = "{:X}".format(i)
        n_binary = "{:b}".format(i)
        print(n_decimal.rjust(width),n_octal.rjust(width),n_hex.rjust(width),n_binary.rjust(width))
        
if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

```python
#Alphabet Rangoli

import string

alpha = string.ascii_lowercase
def print_rangoli(n):
    L = []
    for i in range(n):
        s = "-".join(alpha[i:n])
        L.append((s[::-1]+s[1:]).center(4*n-3, "-"))
    print('\n'.join(L[::-1]+L[1::]))

f __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
#The Minion Game

def minion_game(string):
    vowels = ['A','E','I','O','U']
    k = 0
    s = 0
    for i in range(len(string)):
        if string[i] in vowels:
            k += (len(string)-i)
        else:
            s += (len(string)-i)
    if k > s:
        print ("Kevin", k)
    elif k < s:
        print ("Stuart", s)
    else:
        print ("Draw")
        
if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
#Merge the Tools!

def merge_the_tools(string, k):
    for i in range(0,len(string),k):
        s_s = string[i:i+k]
        l = []
        g = []
        for elem in s_s:
            l.append(elem)
        for letter in l:
            if letter not in g:
                    g.append(letter)
        print("".join(g))
        g = []

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

```python
#Collections.deque()

from collections import deque
n = int(input())
d = deque()
for i in range(n):
    op = input().split()
    if len(op) > 1:
        op[1] = int(op[1])
    if op[0] == 'append':
        d.append(op[1])
    if op[0] == 'pop':
        d.pop()
    if op[0] == 'appendleft':
        d.appendleft(op[1])
    if op[0] == 'popleft':
        d.popleft()
for elem in d:
    print(elem,end=' ')
```

```python
#Time Delta

from datetime import datetime
import math
import os
import random
import re
import sys

def time_delta(t1, t2):
    first = datetime.strptime(t1,'%a %d %b %Y %H:%M:%S %z')
    second = datetime.strptime(t2,'%a %d %b %Y %H:%M:%S %z')
    return str(abs(int((first-second).total_seconds())))


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    t = int(input())
    for t_itr in range(t):
        t1 = input()
        t2 = input()
        delta = time_delta(t1, t2)
        fptr.write(delta + '\n')
    fptr.close()
```

```python
#HTML Parser - Part 1

from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):        
        print ('Start :',tag)
        for ele in attrs:
            print ('->',ele[0],'>',ele[1])
            
    def handle_endtag(self, tag):
        print ('End   :',tag)
        
    def handle_startendtag(self, tag, attrs):
        print ('Empty :',tag)
        for ele in attrs:
            print ('->',ele[0],'>',ele[1])
            
MyParser = MyHTMLParser()
MyParser.feed(''.join([input().strip() for _ in range(int(input()))]))
```

```python
#HTML Parser - Part 2

from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if '\n' in data:
            print ('>>> Multi-line Comment')
        else:
            print ('>>> Single-line Comment')
        print(data)
    def handle_data(self, data):
        if data != '\n':
            print ('>>> Data')
            print(data)

html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#Validating UID

import re
t = int(input())
for i in range(t):
    uid = input()
    if (len(uid) != 10 or not uid.isalnum or len(uid) != len(set(uid)) or
    len(re.findall(r'[0-9]', uid)) < 3 or len(re.findall(r'[A-Z]', uid)) < 2):
        print('Invalid')
    else:
        print('Valid')
```

```python
#Validating Credit Card Numbers

import re
n = int(input())
for i in range(n):
    ccn = input()
    if (re.fullmatch(r'[4-6]\d{3}-?\d{4}\-?\d{4}\-?\d{4}',ccn) and not (bool(re.search(r'([\d])\1\1\1', ccn.replace("-",''))))):
        print('Valid')
    else:
        print('Invalid')
```

```python
###XML 
#XML 1 - Find the Score

import sys
import xml.etree.ElementTree as etree

def get_attr_number(node):
    ntot = 0
    for elem in node.iter():
        ntot += len(elem.attrib)
    return ntot

if __name__ == '__main__':
    sys.stdin.readline()
    xml = sys.stdin.read()
    tree = etree.ElementTree(etree.fromstring(xml))
    root = tree.getroot()
    print(get_attr_number(root))
```

```python
#XML2 - Find the Maximum Depth

import xml.etree.ElementTree as etree

maxdepth = 0
def depth(elem, level):
    global maxdepth
    level += 1
    maxdepth = max(maxdepth, level)
    for c in elem:
        depth(c,level)

if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

```python
###Number Line Jumps

import math
import os
import random
import re
import sys

def kangaroo(x1, v1, x2, v2):
    default = 'NO'
    for i in range(10000):
        if x1 + v1*i == x2 + v2*i:
            default = 'YES'
    return(default)
            
        

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    x1V1X2V2 = input().split()
    x1 = int(x1V1X2V2[0])
    v1 = int(x1V1X2V2[1])
    x2 = int(x1V1X2V2[2])
    v2 = int(x1V1X2V2[3])
    result = kangaroo(x1, v1, x2, v2)
    fptr.write(result + '\n')
    fptr.close()
```

```python
###Viral Advertising

import math
import os
import random
import re
import sys


def viralAdvertising(n):
    sh = 5
    cm = 0
    for i in range (n):
        l = sh//2
        cm += l
        sh = l*3
    return cm

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    n = int(input())
    result = viralAdvertising(n)
    fptr.write(str(result) + '\n')
    fptr.close()

```

```python
###Recursive Digit Sum

import math
import os
import random
import re
import sys

def superDigit(n, k):
    l = []
    for i in range(len(n)):
        l.append(int(n[i]))
    sm = sum(l)*k
    st = str(sm)
    if len(st) > 1:
        return superDigit(st, 1)
    else:
        return sm

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')
    nk = input().split()
    n = nk[0]
    k = int(nk[1])
    result = superDigit(n, k)
    fptr.write(str(result) + '\n')
    fptr.close()
```

```python
###Insertion Sort - Part 1

import math
import os
import random
import re
import sys

def insertionSort1(n, arr):
    i = n-1
    nto = arr[i]
    while(arr[i-1] > nto and i > 0):
        arr[i] = arr[i-1]
        print (' '.join(map(str,arr)))
        i -= 1
    arr[i] = nto
    print (' '.join(map(str,arr)))
    return arr

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().rstrip().split()))
    insertionSort1(n, arr)
```
